### **Metric Evaluation**

#### **Cancellation Prediction**
Untuk menilai seberapa baik model dalam memprediksi target, diperlukan evaluasi menggunakan metrik yang mengukur tingkat error. Fokus utama analisis ini adalah meningkatkan pendapatan dengan mengoptimalkan model pada kelas positif, yaitu dengan meminimalkan kerugian akibat pemesanan hotel yang sebenarnya dibatalkan namun diprediksi tidak batal (False Negative), maupun pemesanan yang sebenarnya tidak batal namun diprediksi batal (False Positive). Pada studi ini, target diklasifikasikan sebagai berikut:
- 0 : Not Canceled
- 1 : Canceled

<br></br>
![Confusion Metric](https://media.springernature.com/full/springer-static/image/art%3A10.1038%2Fnmeth.3945/MediaObjects/41592_2016_Article_BFnmeth3945_Fig1_HTML.jpg)

| **ACTUAL** | **PREDICTED** |  |
|------------|---------------|--|
|            | **Not Canceled (0)** | **Canceled (1)** |
| **Not Canceled (0)** | **TRUE NEGATIVE (TN)**<br>Model memprediksi pemesanan tidak dibatalkan dan memang tidak dibatalkan. | <span style="color:red;"><b>FALSE POSITIVE (FP)</b></span><br><span style="color:red;">Model memprediksi booking akan dibatalkan (*canceled*), padahal sebenarnya tidak dibatalkan.</span> |
| **Canceled (1)** | <span style="color:red;"><b>FALSE NEGATIVE (FN)</b></span><br><span style="color:red;">Model memprediksi booking tidak akan dibatalkan, padahal sebenarnya dibatalkan.</span> | **TRUE POSITIVE (TP)**<br>Model memprediksi pemesanan dibatalkan dan memang dibatalkan. |


Kondisi FP dan FN dapat berimplikasi sebagai berikut:
- **Type 1 Error False Positive (FP):** Hotel mungkin menolak atau mengantisipasi booking yang sebenarnya valid, sehingga berpotensi kehilangan pendapatan dari tamu yang seharusnya datang. Kondisi ini juga dapat menyebabkan double booking yang akhirnya akan menolak salah satu tamu, sehingga akan berdampak pada review negatif dan membutuhkan biaya untuk mengembalikan citra positif.
- **Type 2 Error False Negative (FN):** Hotel tidak mengantisipasi pembatalan yang sebenarnya terjadi, sehingga kamar yang sudah disiapkan menjadi kosong dan berpotensi menyebabkan *opportunity loss*.


Selanjutnya, akan dilakukan perhitungan cost atau kerugian yang ditimbulkan dari masing-masing jenis kesalahan prediksi tersebut. Analisis ini bertujuan untuk mengetahui mana di antara FP dan FN yang memberikan dampak lebih signifikan terhadap bisnis hotel, sehingga dapat digunakan sebagai dasar dalam menentukan metric evaluasi yang paling sesuai untuk model klasifikasi yang dikembangkan.

##### **Asumsi yang digunakan:**

- ADR (Average Daily Rate): rata-rata harga jual kamar per malam. Asumsi: 104 EUR (mengikuti rata-rata dataset)
- Biaya variabel/room: biaya yang hanya keluar jika kamar terisi (linen, amenity, utilitas). (Asumsi: 30 EUR)
- Margin per malam: pendapatan kontribusi = ADR − biaya variabel. (104 − 30 = 74 EUR)
- Cancellation fee %: porsi ADR yang tetap ditagih saat tamu membatalkan. (Asumsi: 20% ADR = 21 EUR)
- Resale probability (resale\_prob): peluang kamar bisa terjual lagi last-minute setelah ada pembatalan.
- Resale ADR factor: diskon harga saat terjual ulang (mis. 0.8 × ADR berarti jual 80% dari harga normal).
- Prep waste: biaya kecil yang sudah terlanjur keluar saat no-show (penataan kamar, admin). (Asumsi: 8 EUR)
- Walk/p\_walk: risiko hotel harus memindahkan tamu ke hotel lain (overbook) beserta pembiayaannya.
- Walk cost: biaya kompensasi untuk “walked guest” (akomodasi lain, transport, voucher). (Asumsi: 180 EUR)
- p\_bad\_review: peluang dampak reputasi/ulasan buruk/churn karena pengalaman buruk.
- LTV\_loss: kehilangan nilai pelanggan masa depan (lifetime value) karena kecewa & tidak kembali. ( Asumsi: 130–250 EUR)
- Marketing recovery: biaya pemulihan citra/ops per insiden (gestur goodwill, follow-up, dll). (Asumsi: 12 EUR)
- Ad Spend: biaya iklan pemulihan citra (mis. kampanye review-posi­tif, voucher, social ads)

##### **Biaya base case:**

- ADR = €104, biaya variabel = €30 ⇒ margin = €74
- Cancellation fee = 20% × €104 = €20.8 ≈ €21
- Resale prob = 0.4 (40% kamar bisa dijual ulang last-minute)
- Resale ADR = 0.8 × €104 = €83.2 ⇒ resale margin = €53.2 (≈ €53)
- Prep waste = €8
- p_walk = 0.1, walk cost = €180
- p_bad_review = 0.3, LTV_loss = €200 (pakai nilai tengah dari rentang €130–€250; bisa diganti sesuai preferensi)
- Marketing recovery = €12
- Ad Spend (FP) = €100

##### **Rumus biaya (per 1 kejadian salah prediksi)**

1. False Negative (FN). Prediksi tidak cancel, faktanya cancel → kamar berpotensi kosong, tetapi sebagian bisa ditutup oleh cancellation fee & resale.

    Cost_FN = ((1 − resale_prob) × margin) − cancellation_fee  + prep_waste − (resale_prob × resale_margin)

    = ((1 − 0.4) × 74) − 20.8 + 8 − (0.4 × 53.2) 

    = 44.4 − 20.8 + 8 − 21.28

    = €10.32

2. False Positive (FP). Prediksi cancel, faktanya hadir/valid → kehilangan margin tamu valid, jika overbook bisa berujung “walk”, dan terdapat dampak reputasi.
    Cost_FP

    = lost_margin + (p_walk × walk_cost) + (p_bad_review × LTV_loss) + marketing_recovery + Ad spend per incident

    = 74 + (0.1 × 180) + (0.3 × 200) + 12 + 100                

    = 74 + 18 + 60 + 12 + 100

    = €264


**Kesimpulan base case:**
FP = €264 per kejadian vs FN = $10.32

Karena FP lebih merugikan, maka:

- Tekan FP -> akan digunakan metric evaluation **Precision**. Supaya “yang kita sebut cancel” benar-benar berpotensi cancel.
- **F0.5 (β < 1)** agar pro-precision .
- Setelah pemodelan, akan dihitung expected cost per N reservasi: (#FP × €264) + (#FN × €10.32) dari confusion matrix validasi.


##### Rumus Precision dan F-beta Score (β=0.5):

- **Precision:**  
    $$ \text{Precision} = \frac{TP}{TP + FP} $$

- **F-beta Score (β=0.5):**  
    $$ F_{0.5} = \frac{(1 + 0.5^2) \cdot (\text{Precision} \cdot \text{Recall})}{(0.5^2 \cdot \text{Precision}) + \text{Recall}} $$

Keterangan:  
- **TP** = True Positive  
- **FP** = False Positive  
- **Recall** = $$ \frac{TP}{TP + FN} $$

### **`Distribusi Kolom Numerik`**

```python
from scipy.stats import normaltest

green_pink = ['#43a047', '#e75480']

for idx in range(0, len(num_features), 2):
    cols = num_features[idx:idx+2]
    ncols = len(cols)
    fig, axes = plt.subplots(
        2, ncols, figsize=(7*ncols, 5), 
        gridspec_kw={"height_ratios": (.18, .82)}, 
        constrained_layout=True  # ganti tight_layout
    )
    if ncols == 1:
        axes = axes.reshape(2, 1)
    for j, col in enumerate(cols):
        data = df_eda1[col].dropna()
        stat, pval = normaltest(data)
        norm_label = f"p-value = {pval:.5f}, {'Normal' if pval > 0.05 else 'TIDAK Normal'}"

        # Boxplot
        sns.boxplot(x=data, ax=axes[0, j], color=green_pink[1], medianprops=dict(color="orange"))
        axes[0, j].axvline(data.mean(), color="red", ls="--")
        axes[0, j].set_title(f"Distribusi {col}", fontsize=13, fontweight="bold", pad=25)  # tambah jarak
        axes[0, j].text(0.5, 1.05, norm_label, ha="center", va="bottom", transform=axes[0, j].transAxes)

        # Histogram
        sns.histplot(data, ax=axes[1, j], kde=True, bins=50, color=green_pink[0], stat="count")
        axes[1, j].axvline(data.mean(), color="red", ls="--")
        axes[1, j].set_xlabel(col)
        axes[1, j].set_ylabel("Frequency")
    
    plt.show()
```

## **3.3 Data Cardinalities**

Analisis ini penting untuk memahami keragaman data, mengidentifikasi fitur dengan kategori sangat banyak (high cardinality), serta menentukan strategi encoding yang tepat pada tahap preprocessing. Fitur dengan cardinality tinggi dapat mempengaruhi kompleksitas model dan membutuhkan penanganan khusus agar proses pemodelan tetap efisien dan akurat.

```python
green_pink = ["#43a047", "#e75480", "#66bb6a", "#ec8aa4", "#a5d6a7", "#f3c2cf"]

for col in cat_features:
    plt.figure(figsize=(12, 6))
    n_unique = df_eda1[col].nunique(dropna=False)
    order = df_eda1[col].value_counts().index
    
    if n_unique > 10:
        top10 = df_eda1[col].value_counts().nlargest(10)
        order = top10.index
        data_plot = df_eda1[df_eda1[col].isin(order)]
        total = len(data_plot)
        title = f"Cardinality of {col} (Top 10)"
        ax = sns.countplot(data=data_plot, x=col, order=order, palette=green_pink)
    else:
        total = len(df_eda1)
        title = f"Cardinality of {col}"
        ax = sns.countplot(data=df_eda1, x=col, order=order, palette=green_pink)
    
    for p in ax.patches:
        count = int(p.get_height())
        percent = 100 * count / total
        ax.annotate(f"{count:,}\n({percent:.1f}%)",
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='bottom',
                    fontsize=9, color='black',
                    xytext=(0, 8), textcoords='offset points')
    
    ax.set_title(title, fontsize=14, fontweight="bold", pad=20)
    ax.set_xlabel(col)
    ax.set_ylabel("Count")
    plt.xticks(rotation=60, ha='right')
    plt.tight_layout()
    plt.show()
```

### **Identify Outlier & Anomaly**

Identifikasi outlier dan data anomali merupakan langkah penting dalam proses analisis data karena keberadaan nilai-nilai ekstrem atau tidak wajar dapat memengaruhi hasil statistik, visualisasi, dan performa model machine learning. Outlier dapat menyebabkan bias pada estimasi parameter, menurunkan akurasi prediksi, serta menyesatkan interpretasi pola data. Dengan mendeteksi dan menangani outlier serta anomali secara tepat, kualitas data menjadi lebih baik sehingga hasil analisis dan model yang dibangun menjadi lebih andal dan representatif terhadap kondisi sebenarnya.

```python
# 1. ADR negatif (< 0)
anom_adr_neg = df_eda1[df_eda1['adr'] < 0]

# 2. ADR = 0 padahal ada tamu yang menginap dan status checkout
anom_adr_zero = df_eda1[
    (df_eda1['adr'] == 0) &
    ((df_eda1['adults'] + df_eda1['children'] + df_eda1['babies']) > 0) &
    (df_eda1['reservation_status'] == 'Check-Out')
]

# 3. ADR sangat tinggi (misal: 5000 atau lebih)
anom_adr_high = df_eda1[df_eda1['adr'] >= 5000]

# 4. Jumlah tamu (adults + children + babies) = 0 padahal status reservasi checkout
anom_zero_guest = df_eda1[
    ((df_eda1['adults'] + df_eda1['children'] + df_eda1['babies']) == 0) &
    (df_eda1['reservation_status'] == 'Check-Out')
]

# 5. Durasi menginap (stays_in_week_nights + stays_in_weekend_nights) = 0 padahal status reservasi checkout
anom_zero_stay = df_eda1[
    ((df_eda1['stays_in_week_nights'] + df_eda1['stays_in_weekend_nights']) == 0) &
    (df_eda1['reservation_status'] == 'Check-Out')
]

# Anomali
print("=== Anomali Outlier ===")
print("ADR negatif:", len(anom_adr_neg))
print("ADR = 0 (ada tamu, status checkout):", len(anom_adr_zero))
print("ADR >= 5000:", len(anom_adr_high))
print("Jumlah tamu = 0 padahal status Checkout:", len(anom_zero_guest))
print("Durasi menginap = 0 padahal status Checkout:", len(anom_zero_stay))
```

Berdasarkan identifikasi **anomali** pada data hotel booking, ditemukan beberapa kasus yang perlu perhatian khusus:
- Terdapat data dengan nilai ADR (Average Daily Rate) negatif maupun sangat tinggi (≥ 5000), yang secara logika tidak mungkin terjadi dalam transaksi hotel dan kemungkinan besar merupakan kesalahan input atau pencatatan.
- Ditemukan juga pemesanan dengan ADR = 0, padahal ada tamu yang menginap dan status reservasi sudah check-out. Hal ini mengindikasikan adanya transaksi yang tidak tercatat dengan benar atau kemungkinan promosi khusus yang tidak terdokumentasi.
- Selain itu, terdapat pemesanan dengan jumlah tamu (dewasa + anak + bayi) = 0 namun status reservasi check-out, serta durasi menginap (stays_in_week_nights + stays_in_weekend_nights) = 0 pada status check-out. Kedua kondisi ini tidak logis karena tidak mungkin ada reservasi tanpa tamu atau tanpa menginap yang berstatus check-out.
- Anomali-anomali tersebut perlu ditangani pada tahap data cleaning, misalnya dengan menghapus atau memperbaiki data terkait, agar tidak memengaruhi hasil analisis dan pemodelan selanjutnya.

Identifikasi **outlier** pada fitur numerik sudah terlihat jelas pada bagian analisis distribusi data, terutama melalui visualisasi boxplot dan histogram. Outlier semacam ini umumnya masih wajar dan tidak selalu merupakan kesalahan data.

Namun, untuk memastikan model machine learning tetap robust, outlier-outlier tersebut **dapat diatasi dengan** beberapa pendekatan, seperti menggunakan **teknik** **scaling** yang tidak sensitif terhadap nilai ekstrem (misal: RobustScaler) atau memilih **algoritma** yang memang tahan terhadap outlier, seperti tree-based models (Decision Tree, Random Forest, XGBoost). Anomali biasanya merupakan hasil kesalahan input, pencatatan, atau kasus yang tidak logis secara bisnis. Jika dibiarkan, anomali ini dapat menyebabkan:

- **Model menjadi bias:** Nilai ekstrem atau tidak logis dapat mengacaukan distribusi data, sehingga model belajar dari pola yang tidak representatif.
- **Penurunan akurasi:** Model bisa salah dalam mengidentifikasi pola penting, terutama pada fitur numerik seperti ADR, jumlah tamu, atau durasi menginap.
- **Generalizability buruk:** Model yang dilatih dengan data anomali cenderung overfitting pada noise, sehingga performanya menurun saat memprediksi data baru di masa depan.

Dengan menghapus anomali, model akan belajar dari data yang lebih bersih dan realistis, sehingga hasil prediksi lebih stabil, akurat, dan dapat diandalkan untuk kasus nyata di masa depan. Ini meningkatkan kemampuan model dalam mengidentifikasi pola pembatalan yang benar-benar relevan secara bisnis. Oleh karena itu penghapusan data anomali sangat penting untuk memastikan kualitas data. 

Pada kasus hotel booking ini, anomali yang dapat terjadi yaitu:
1. Nilai ADR (Average Daily Rate) negatif dan sangat tinggi. 
    - Threshold 0-600 diputuskan berdasarkan distribusi data, dimana terdapat 1 nilai ADR ekstrem (5000 Euro) padahal pada kepadatan sebaran data terlihat bahwa cluster nilai ADR tertinggi tidak mencapai 600 Euro. Jika rata-rata ADR adalah 106, maka ADR lebih dari 5000, yang merupakan nilai paling tinggi ADR pada dataset ini tidaklah logis. Begitu pula dengan ADR negatif. 
2. Reservasi dengan tarif harian rata-rata (ADR) 0 walaupun ada tamu yang menginap dan sudah berstatus 'Check Out'. 
3. Reservasi dimana ada tamu yang menginap dan berstatus 'Check Out'. 
4. Reservasi dimana durasi menginap adalah 0 (nol) dan berstatus 'Check Out'. 


```python
# Handling anomaly pada df sesuai temuan dari tahapan sebelumnya
print("Jumlah data sebelum handling anomaly:", len(df))

#1. Remove bookings with illogical data and filter ADR to a reasonable range.
initial_rows = df.shape[0]
df = df[df['adr'] >= 0]
df = df[df['adr'] <= 600]

print(f"\nRemoved {initial_rows - df.shape[0]} rows with ADR outside the 0-600 range.")
# Bookings must have at least one guest.
initial_rows = df.shape[0]


# 2. Hapus baris dengan ADR = 0, ada tamu menginap, dan status checkout
df = df[~(
    (df['adr'] == 0) &
    ((df['adults'] + df['children'] + df['babies']) > 0) &
    (df['reservation_status'] == 'Check-Out')
)]

# 3. Hapus baris dengan ADR sangat tinggi (>= 5000)
df = df[df['adr'] < 5000]

# 4. Hapus baris dengan jumlah tamu = 0 dan status checkout
df = df[~(
    ((df['adults'] + df['children'] + df['babies']) == 0) &
    (df['reservation_status'] == 'Check-Out')
)]

# 5. Hapus baris dengan durasi menginap = 0 dan status checkout
df = df[~(
    ((df['stays_in_week_nights'] + df['stays_in_weekend_nights']) == 0) &
    (df['reservation_status'] == 'Check-Out')
)]

print("Sisa data setelah handling anomaly:", len(df))
```

#### **Outlier**
Jumlah outlier pada fitur numerik cukup banyak dan ternyata memang mewakili karakteristik mayoritas data (misalnya karena distribusi data yang memang skewed ), maka outlier tersebut tidak dihapus agar tidak menghilangkan informasi penting. Sebagai solusinya, kita dapat menggunakan **teknik scaling yang robust terhadap outlier**, seperti RobustScaler, atau memilih **algoritma machine learning yang tidak sensitif terhadap nilai ekstrem**, seperti tree-based models (Decision Tree, Random Forest, XGBoost). Dengan demikian, model tetap dapat belajar dari seluruh data tanpa terpengaruh secara signifikan oleh outlier.

### **Drop Columns**

Pada tahap ini kita akan mengeluarkan kolom-kolom atribut dataset yang tidak diperlukan untuk melatih model machine learning.

### **Scaling**

Scaling dilakukan untuk mengubah data-data numerikal agar memiliki skala yang sama. Hal ini dilakukan sebagai solusi algoritma machine learning yang dapat memberikan bobot yang berbeda pada fitur yang memiliki skala berbeda. 

Pada tahap data understanding dan EDA kita menemukan bahwa kolom-kolom numerikal pada dataset memiliki skala (jarak angka minimal & maksimal) yang beragam. Sehingga scaling perlu dilakukan. Selain itu, kita juga mengetahui bahwa hampir seluruh kolom/atribut memiliki banyak data outlier. Sehingga metode scaling yang sesuai untuk digunakan pada dataset ini adalah `RobustScaler` yang tidak sensitif terhadap data outlier. 

### **Encoding**

Encoding adalah proses mengubah data ke dalam format lain. Encoding perlu dilakukan untuk training dataset machine learning untuk merepresentasikan nilai variabel kategorikal agar dapat digunakan oleh algoritma machine learning.  

Dalam dataset ini tidak ditemukan variabel yang memiliki sifat ordinal (memiliki bobot berbeda tiap kategorinya (terkait dengan pengaruhnya dengan target yang hendak diprediksi)). Selain itu, pada kolom/variabel yang akan digunakan untuk pemodelan machine learning juga tidak ditemukan variabel dengan variasi yang sangat banyak (yang memerlukan binary encoding). Sehingga metode encoding yang paling sesuai dan cukup untuk fitur-fitur dataset adalah `One Hot Encoding`.

### Model Benchmarking & Cross Validation

```python
# Define feature lists for the preprocessor.
numerical_features = X.select_dtypes(include=np.number).columns.tolist()
# The room types are now included here as categorical features
categorical_features = X.select_dtypes(include='object').columns.tolist()

# Define the preprocessing pipeline to handle NaNs and scale numerical features.
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', RobustScaler())
])

# Define the preprocessing pipeline with one-hot encoding for categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features)
    ],
    remainder='passthrough'
)

# Define the models to be tested.
models = {
    'Logistic Regression': LogisticRegression(random_state=888, solver='liblinear'),
    'Decision Tree': DecisionTreeClassifier(random_state=888),
    'Random Forest': RandomForestClassifier(random_state=888),
    'Gradient Boost': GradientBoostingClassifier(random_state=888),
    'XGBoost': XGBClassifier(random_state=888, eval_metric='logloss'),
}

# The metrics remain the same
f0_5_scorer = make_scorer(fbeta_score, beta=0.5)
pr_auc_scorer = make_scorer(average_precision_score)

scoring_metrics = {
    'Accuracy': 'accuracy',
    'F1': 'f1',
    'F0.5': f0_5_scorer,
    'Precision': 'precision',
    'Recall': 'recall',
    'ROC AUC': 'roc_auc',
    'PR AUC': pr_auc_scorer
}

results = []
for name, model in models.items():
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])
    
    scores = cross_validate(pipeline, X_train, y_train, cv=5, scoring=scoring_metrics)
    
    # Store the average scores for each model.
    model_results = {'Model': name}
    for metric_name, score_values in scores.items():
        if metric_name.startswith('test_'):
            model_results[metric_name.replace('test_', '')] = score_values.mean()
    
    results.append(model_results)

df_scores = pd.DataFrame(results)
print("\nModel Performance Comparison (Cross-Validation):")
print(df_scores)
```

### Hyperparameter Tuning

Hyperparameter tuning bertujuan untuk mengoptimalkan performa model dengan mencari kombinasi parameter terbaik. Proses ini dilakukan pada model **XGBoost** yang sebelumnya terbukti memberikan hasil prediksi paling baik dari seluruh kandidat model.

XGBoost (Extreme Gradient Boosting) adalah algoritma ensemble berbasis decision tree yang sangat populer karena kecepatan dan akurasinya. XGBoost membangun model secara **bertahap/sequential** dengan menambahkan pohon-pohon baru yang fokus memperbaiki kesalahan prediksi dari pohon-pohon sebelumnya. Setiap iterasi, model baru berusaha meminimalkan error dengan teknik gradient descent pada fungsi loss yang digunakan.

Proses XGBoost secara umum:
1. Model pertama dibuat untuk memprediksi target.
2. Model berikutnya dilatih untuk memprediksi sisa error (residual) dari model sebelumnya.
3. Proses ini diulang berkali-kali, sehingga setiap pohon baru "mengoreksi" kesalahan model sebelumnya.
4. Prediksi akhir merupakan gabungan (penjumlahan) dari seluruh pohon dengan bobot tertentu.

XGBoost juga memiliki fitur regularisasi yang kuat untuk mencegah overfitting dan mendukung berbagai parameter yang dapat diatur untuk mengoptimalkan performa.

**Parameter pada Hyperparameter Tuning XGBoost:**

- n_estimators: Jumlah pohon (tree) yang akan dibangun secara berurutan. Semakin banyak, model bisa lebih kompleks, namun risiko overfitting meningkat.
- max_depth: Kedalaman maksimum tiap pohon. Mengontrol kompleksitas pohon; pohon yang lebih dalam dapat menangkap pola lebih rumit, tapi juga lebih mudah overfit.
- learning_rate: Bobot kontribusi setiap pohon baru terhadap prediksi akhir. Nilai kecil membuat proses belajar lebih lambat namun bisa meningkatkan akurasi (dengan n_estimators lebih banyak).
- subsample: Proporsi data latih yang digunakan untuk membangun setiap pohon. Nilai <1 membuat model lebih robust terhadap overfitting.
- colsample_bytree: Proporsi fitur yang digunakan untuk membangun setiap pohon, membantu mengurangi korelasi antar pohon.
- gamma: Minimum loss reduction yang diperlukan agar node dibagi. Semakin besar nilai gamma, pohon akan lebih konservatif.
- scale_pos_weight: Bobot untuk menangani ketidakseimbangan kelas pada data target.

Dengan tuning parameter-parameter ini, kita mencari kombinasi yang menghasilkan model paling akurat dan generalisasi terbaik pada data baru.

Hyperparameter Tuning dilakukan untuk mencari kombinasi parameter yang menghasilkan model paling akurat dan generalisasi terbaik pada data baru. Proses ini biasanya dilakukan menggunakan teknik seperti Grid Search atau Randomized Search untuk mengeksplorasi ruang parameter yang telah ditentukan.

Setiap kombinasi parameter akan dievaluasi dengan data validasi atau melalui cross-validation untuk memastikan bahwa peningkatan performa bukan disebabkan oleh overfitting pada data training.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier

# Redefine the pipeline to include SMOTE
xgboost_model = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(random_state=888, use_label_encoder=False, eval_metric='logloss'))
])

# Calculate the ratio of negative to positive classes for XGBoost
scale_pos_weight_value = (y_train == 0).sum() / (y_train == 1).sum()
print(f"\nScale_pos_weight value for XGBoost: {scale_pos_weight_value:.2f}")
# This tells the model "the negative class is X times larger than the positive class"

# Define the parameter space for XGBoost
xgboost_param_dist = {
    'classifier__n_estimators': np.arange(100, 1001, 100),
    'classifier__learning_rate': np.linspace(0.01, 0.3, 10),
    'classifier__max_depth': np.arange(3, 15, 1), # Increased max range
    'classifier__subsample': np.linspace(0.6, 1.0, 5),
    'classifier__colsample_bytree': np.linspace(0.6, 1.0, 5),
    'classifier__gamma': [0, 0.1, 0.2, 0.5],
    'classifier__scale_pos_weight': [1, scale_pos_weight_value, 10] # <-- CRITICAL ADDITION
}

random_search = RandomizedSearchCV(
    estimator=xgboost_model,
    param_distributions=xgboost_param_dist,
    n_iter=50,
    cv=5,
    scoring=f0_5_scorer,
    random_state=888,
    n_jobs=-1,
    verbose=1
)

random_search.fit(X_train, y_train)

print("\nBest Hyperparameters found for XGBoost Classifier:")
print(random_search.best_params_)
print("\nBest F0.5 Score on training data:", random_search.best_score_)
```

Setelah proses tuning selesai, kita mendapatkan satu set kombinasi hyperparameter terbaik (best parameters) yang memungkinkan model XGBoost mencapai F0.5 score tertinggi, yaitu 0.7259. 

Model akhir yang telah di-tuning ini siap untuk diuji pada dataset testing yang belum pernah dilihat sebelumnya. Pengujian ini bertujuan untuk memberikan gambaran yang unbiased tentang seberapa baik model akan performa pada data dunia nyata. Dengan demikian, hyperparameter tuning adalah langkah kritis untuk mengubah model yang "cukup baik" menjadi model yang optimal dan siap produksi.

### **SHAP**
**SHAP** adalah metode interpretasi model berbasis teori permainan Shapley Value dari matematika ekonomi. Untuk satu prediksi, SHAP menghitung berapa besar kontribusi masing-masing fitur jika dibandingkan dengan prediksi rata-rata (baseline). Tidak hanya menunjukkan seberapa sering atau seberapa besar fitur digunakan dalam tree tetapi juga menjelaskan arah dan konteks kontribusinya. SHAP (SHapley Additive exPlanations) memberikan penjelasan yang lebih lengkap dan intuitif terhadap kontribusi setiap fitur terhadap prediksi individu, arah pengaruh fitur (menaikkan atau menurunkan prediksi), dan pentingnya fitur secara global dan lokal (per data point).

```python
import shap

# Get the preprocessor and classifier from best_model pipeline
preprocessor = best_model.named_steps['preprocessor']
classifier = best_model.named_steps['classifier']

# Get the feature names from the preprocessor
feature_names = preprocessor.get_feature_names_out()

# Transform the original X_test data using the preprocessor
X_test_transformed = preprocessor.transform(X_test)
X_test_transformed_df = pd.DataFrame(X_test_transformed, columns=feature_names)

# Create the SHAP explainer with the trained classifier
explainer = shap.Explainer(classifier)

# Compute SHAP values on the transformed test data
shap_values = explainer(X_test_transformed_df)
```

Dalam interpretasi model machine learning menggunakan SHAP, terdapat dua konsep penting: **global feature importance** dan **feature impact direction**. 
Global feature importance menggambarkan seberapa besar kontribusi rata-rata setiap fitur terhadap prediksi model secara keseluruhan, sehingga kita dapat mengetahui fitur mana yang paling berpengaruh dalam menentukan hasil prediksi. 
Global Feature Importance (SHAP):
-   Menampilkan rata-rata absolut nilai SHAP untuk setiap fitur.
-   Mengukur seberapa besar kontribusi rata-rata fitur terhadap prediksi model (tanpa memperhatikan arah pengaruh).
-   Semakin panjang bar, semakin penting fitur tersebut secara global.
-   Tidak menunjukkan apakah fitur menaikkan atau menurunkan prediksi.
